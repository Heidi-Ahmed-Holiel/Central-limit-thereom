```python
import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns
```

```python
#assuming 
rate = 0.25
#Population mean is going to be
mu = 1/rate

# to generate exponential distribution you can use 
random_samples=np.random.exponential((1/rate), 10000)

```

```python
def simulate_central_limit(random_samples, sample_size,n_samples):
  #assuming 
  rate = 0.25
  #Population mean is going to be
  mu = 1/rate

  means = []
  samples = []
  avgs=[]
  var=[]
  
  for i in range(0, n_samples):
    samples.append(np.random.choice(random_samples, sample_size, n_samples))
    avgs.append(np.mean(samples[i]))
  mean=(np.mean(avgs))
  var=(np.var(avgs))

  return avgs, mean, var

```

```python
#First exp
# start with 50 random samples, sample size = 2 

avgs1, mean1, var1 = simulate_central_limit(random_samples,2, 50)


#Second exp
# use another 50 random samples ,sample size= 500
avgs2, mean2, var2 = simulate_central_limit(random_samples,500, 50)

#Third exp
# use another 50 random samples ,sample size= 1000
avgs3, mean3, var3 = simulate_central_limit(random_samples,1000, 50)

#Fourth exp
# use another 100 random samples ,sample size= 1000
avgs4, mean4, var4 = simulate_central_limit(random_samples,1000, 100)

#Fifth exp
# use another 1000 random samples ,sample size= 1000
avgs5, mean5, var5 = simulate_central_limit(random_samples,1000, 1000)

#Sixth exp
# use another 5000 random samples ,sample size= 1000
avgs6, mean6, var6 = simulate_central_limit(random_samples,1000, 5000)

print("First 50 random samples, sample size = 2 : ")
print("Mean = {0:f}, Variance = {1:f}".format(mean1, var1))
print("Average of outcomes in each case: ", avgs1, "\n")

print("Another 50 random samples ,sample size= 500: ")
print("Mean = {0:f}, Variance = {1:f}".format(mean2, var2))
print("Average of outcomes in each case: ", avgs2, "\n")

print("Another 50 random samples ,sample size= 1000: ")
print("Mean = {0:f}, Variance = {1:f}".format(mean3, var3))
print("Average of outcomes in each case: ", avgs3, "\n")

print("Another 100 random samples ,sample size= 1000: ")
print("Mean = {0:f}, Variance = {1:f}".format(mean4, var4))
print("Average of outcomes in each case: ", avgs4, "\n")

print("Another 1000 random samples ,sample size= 1000")
print("Mean = {0:f}, Variance = {1:f}".format(mean5, var5))
print("Average of outcomes in each case: ", avgs5, "\n")

print("Another 5000 random samples ,sample size= 1000: ")
print("Mean = {0:f}, Variance = {1:f}".format(mean6, var6))
print("Average of outcomes in each case: ", avgs6, "\n")
```
